### Data Cleaning.

```python
# Combine column name (first and last) to name column. 
df['name'] = df['first'].str.strip() + ' ' + df['last'].str.strip()
```

```python
df[['first', 'last', 'name']].head()
```

```python
df.columns
```

```python
# Drop useless Columns:

df = df.drop(columns=[
    "Unnamed: 0",
    "first", "last", "street",
    "lat", "long", "merch_lat", "merch_long", "merch_zipcode",
    "trans_num", "unix_time",
    "is_fraud"
])
```

```python
df.columns
```

```python
# Change Data Type of trans_date_trans_time.
df['trans_date_trans_time']=pd.to_datetime(df['trans_date_trans_time'], errors='coerce')
```

```python
# Change data type of dob.
df['dob']=pd.to_datetime(df['dob'], errors='coerce')
```

```python
# Create separate date and time columns.
df['transaction_date'] = df['trans_date_trans_time'].dt.date
df['transaction_time'] = df['trans_date_trans_time'].dt.time
```

```python
df=df.drop(columns=['trans_date_trans_time'])
```

```python
df.columns
```

```python
# Rename column cc_num name to credit_card_num.
df = df.rename(columns={"cc_num": "credit_card_num"})
```

```python
# Removing Invalid Transaction Values.

df=df[df['amt']> 0]
```

```python
# No zero or negative transaction amounts remain after filtering.
(df["amt"] <= 0).sum()
```

```python
# Remove duplicates.

df = df.drop_duplicates()
```

```python
df.shape
```

#### Summary ( Data Cleaning):
Cleaned the dataset by removing irrelevant, misleading, and non-analytical columns, including technical identifiers and unnecessary location details. Filtered out invalid transactions with non-positive amounts and removed duplicate records to maintain data integrity. Corrected data types for date fields, split transaction datetime into separate date and time columns, and standardized column names for clarity. The resulting dataset was structured, consistent, and ready for analysis.

### Feature Engineering.

```python
# Creating age column.
df['age'] = (pd.to_datetime(df['transaction_date']) - pd.to_datetime(df['dob'])).dt.days // 365
```

```python
df['age'].describe()
```

```python
df.columns
```

```python
# Creating a New Customer-Level Table.

customer_df = df.groupby('credit_card_num').agg(total_spend=('amt', 'sum'),
    transaction_count=('amt', 'count'),
    avg_spend=('amt', 'mean'),
    first_transaction=('transaction_date', 'min'),
    last_transaction=('transaction_date', 'max'),
    avg_age=('age', 'mean'),
).reset_index()                   
```

```python
customer_df.head()
```

```python
# Customer Recency (Days Since Last Transaction). (How many days ago did the customer make their last transaction?)

latest_date = pd.to_datetime(customer_df['last_transaction']).max()  #(Most recent transaction date of all customers.)

customer_df['days_since_last_transaction'] = (latest_date - pd.to_datetime(customer_df['last_transaction'])).dt.days  # (how many days ago they made their last transaction, compared to the most recent transaction date in the dataset)
```

```python
customer_df.columns
```

```python
customer_df[["credit_card_num", "last_transaction", "days_since_last_transaction"]].head(10)
```

```python
# Customer Engagement Status (Active vs Inactive).

customer_df['engagement_status'] = customer_df['days_since_last_transaction'].apply(
                                   lambda x: 'Active' if x <= 50 else 'Inactive')     # It means (If a customer made a transaction within the last 50 days --> Active,
                                                                                      # ( If it has been more than 50 days --> Inactive)
```

```python
customer_df[["credit_card_num", "days_since_last_transaction", "engagement_status"]].head(10)
```

```python
# Number of Active vs Inactive Customers.

customer_df["engagement_status"].value_counts()
```
